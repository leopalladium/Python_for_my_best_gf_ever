## Умножение хуй пойми чего на хуй знает что

### Условие

````python
def multiply_numbers(bound: int, numbers: List[int]) -> int:
    """
    Funkce vypočítá součin čísel v seznamu numbers, jejichž hodnota
    je z intervalu 1 až bound (včetně). Pokud se v poli žádná taková
    čísla nenacházejí, vrátí 1.

    časová složitost: O(max(bound, len(numbers))
    prostorová složitost: O(bound)

    :param bound: horní hranice intervalu pro hodnotu čísel, která se
        započítávají do výsledku
    :param numbers: seznam čísel k počítání součinu
    :return: součin čísel v seznamu numbers, jejichž hodnota je z intervalu
        1 až bound (včetně)
    """
    array = [0 for i in range(bound)]
    for i in range(len(numbers)):
        array[numbers[i]] += 1
    val = 1
    for i in range(len(array)):
        for j in range(array[i]):
            val *= i
    return val
````

````python
def multiply_numbers(bound: int, numbers: List[int]) -> int:
    """
    Funkce vypočítá součin čísel v seznamu numbers, jejichž hodnota
    je z intervalu 1 až bound (včetně). Pokud se v poli žádná taková
    čísla nenacházejí, vrátí 1.

    časová složitost: O(max(bound, len(numbers))
    prostorová složitost: O(bound)

    :param bound: horní hranice intervalu pro hodnotu čísel, která se
        započítávají do výsledku
    :param numbers: seznam čísel k počítání součinu
    :return: součin čísel v seznamu numbers, jejichž hodnota je z intervalu
        1 až bound (včetně)
    """
    array = [0 for i in range(bound + 1)] # Добавили + 1, для коррекции на индексацию с 0
    for i in range(len(numbers)):
        if 1 <= numbers[i] <= bound:      # Проверка на не входящее в интервал число
            array[numbers[i]] += 1
    val = 1
    for i in range(len(array)):
        for j in range(array[i]):
            val *= i
    return val
````

## Задачка со скобками

### Условие

````python
def has_correct_parentheses(string: str) -> bool:
    """
    Funkce otestuje, zda zadaný řetězec obsahuje správné ozávorkování,
    tedy před každou uzavírací závorkou musí být příslušná otevírací.
    Řeší se pouze závorky ( ). Vrací True v případě správného
    ozávorkování, jinak False.

    časová složitost: O(len(string))
    prostorová složitost: O(1)

    :param string: testovaný řetězec
    :return: True v případě správného ozávorkování závorkami ( ), jinak False
    """
    opened = 0
    for i in range(len(string)):
        if string[i] == '(':
            opened += 1
        if string[i] == ')':
            opened -= 1
        if opened == 0:
            return True

    return False
````

### Анализ условия
Ф-ция has_correct_parentheses() принимает в себя str и возвращает bool (True | False) в зависимости от того, является ли у нас количество открытых скобок (открывающая скобка без закрывающей или наоборот) нулю. Рассмотрим логику алгоритима. Возвращает *True* если всё ок и открытых нет, иначе *False*.

````python
def has_correct_parentheses(string: str) -> bool: # Декларация, тут всё хорошо
    opened = 0 # Устанавливаем счётчик на ноль, потому что до анализа у нас нету скобок вообще (мы не начали их считать, так что ноль)
    for i in range(len(string)): # Вот тут уже начинаем считать, перебираем каждый символ string от 0 до len(string)
        if string[i] == '(': # Проверяем является ли символ открывающей скобкой
            opened += 1      # есла да, Прибавляем 1 к счётчику, пока мы скобку не закрыли
        if string[i] == ')': # Проверяем является ли символ закрывающей скобкой
            opened -= 1      # если да, Отнимаем 1 от счётчика
        if opened == 0:      # если открытых скобок нет, то
            return True      # ф-ция возвращает True

    return False             # иначе False
````

Буду писать чуть покороче, я тут посмотрел на часы и охуел.
Ошибки которые можно увидеть:
1. Самое понятное это то, что если мы во время каждой итерации проверяем является ли opened == 0, то оно вернёт True даже в случае если мы напишем туда "a(((((((((", где очевидно дохуя незакрытых скобок
2. Если мы запустим функцию с аргументом ")(", то она вернёт нам True, хотя на самом деле скобки не закрыты, это происходит из-за того, что мы считаем только кол-во скобок, но не учитываем их порядок. Если значени opened падает меньше нуля, значит раньше открывающей скобки у нас попалась закрывающая и это сразу False

````python
# Перепишем функцию без ошибок
def has_correct_parentheses(string: str) -> bool:
    opened = 0
    for i in range(len(string)):
        if string[i] == '(':
            opened += 1
        if string[i] == ')':
            opened -= 1
        if opened < 0:    #проверка на закрытость скобок
            return False

    return opened == 0 # вернётся итог сравнения, это сразу проверка на кол-во скобок
````

````python
# Автоматический тест
def ib114_test_brackets() -> None:
    test_cases = [("()", True), (")(", False), ("aaa", True), ("((", False)]

    for i, case in enumerate(test_cases):
        print("Test {} závorkování: vstup {}".format(i + 1, case[0]))
        try:
            if has_correct_parentheses(case[0]) == case[1]:
                print("OK.")
            else:
                print("NOK, {} {} správné uzávorkování a funkce vrátí {}".
                      format(case[0], "je" if case[1] else "není",
                             not case[1]))
        except IndexError as exc:
            print("NOK: přístup mimo pole.")
            print("Chybová hláška Pythonu: {}".format(exc))
ib114_test_brackets()
````

## "Сумма последовательности"

### Условие

````python
def sequence_sum(sequence: List[int]) -> int:
    """
    Funkce spočte "sumu" posloupnosti (sequence) a to tak, že pokud je
    číslo větší než předcházející (sequence[n] > sequence[n-1]), tak ho
    přičte k "sumě", pokud je sequence[n] < sequence[n-1], tak ho odečte
    a pokud je stejné, tak ho přeskočí. První číslo se nezapočítá.

    časová složitost: O(len(sequence))
    prostorová složitost: O(1)

    :param sequence: seznam čísel, pro který se má počítat speciální suma
    :return: speciální suma dle popisu funkce
    """
    strange_sum = 0
    for i in range(len(sequence)):
        if sequence[i] > sequence[i - 1]:
            strange_sum += sequence[i]
        if sequence[i] < sequence[i - 1]:
            strange_sum -= sequence[i]
    return strange_sum
````

Проблема этого условия в том, что ф-ция будет ломаться в момент выполнения if-выражения в строке 16, потому что i начинается с нуля, соответственно `sequence[i - 1]` будет пытаться обратиться к sequence[-1], что невозможно и выбивает ошибку.
Решается легко, нейронка предложила поменять программу так

## последняя задача

````python
def find_substring(string: str, substring: str) -> int:
    """
    Funkce hledá podřetězec (substring) v řetězci (string).
    Pokud se podřetězec v řetězci nachází, vrátí index prvního výskytu.
    Jinak vrací -1.

    časová složitost: O(len(string) * len(substring))
    prostorová složitost: O(1)

    :param string: neprázdný řetězec, jenž se prohledává
    :param substring: neprázdný podřetězec, který se hledá v řetězci 'string'
    :return: Pokud se podřetězec v řetězci nachází, vrátí index prvního
        výskytu. V opačném případě vrací -1.
    """
    if len(substring) > len(string):
        return -1

    j = 1
    i = 1
    while i < len(string):
        if string[i] == substring[j]:
            if j == len(substring) - 1:
                return i - j
            j += 1
    i += 1
    return -1
````

### Анализ условия и разбор полетов

Аня, Клим просил тебе передать, что этот текст уже пишет ИИ, т.к. он невероятно заебался и хочет спать с тобой ️

Эта функция должна найти подстроку (substring) внутри строки (string) и вернуть индекс, с которого она начинается. Сложность должна быть O(n×m) — то есть обычный, прямолинейный поиск перебором без хитрых оптимизаций.

Но тот, кто писал этот код, умудрился собрать в нем самое настоящее бинго из ошибок:

 1. Игнорирование первого символа (Индексация)
    В самом начале мы видим j = 1 и i = 1. Как мы уже помним из предыдущих задач, в Питоне счет начинается с нуля. Начиная с единицы, функция тупо пропускает первые буквы обеих строк.

 2. Бесконечный цикл смерти (Infinite Loop)
    Обрати внимание на отступы: строка i += 1 находится на одном уровне с while i < len(string):, то есть вне самого цикла. Если программа заходит в этот while, она застревает в нем навсегда, потому что i внутри цикла никак не увеличивается (разве что функция чудом наткнется на return). Программа просто зависнет.

3. Отсутствие сброса (Логическая ошибка)
    Представь, что мы ищем слово кот в строке компот. Мы дошли до буквы к, потом о (счетчик j увеличился), а потом бац — буква м вместо т. Что должна сделать нормальная программа? Сбросить j обратно на ноль и начать искать слово кот заново. В этом коде j просто продолжит расти с того места, где остановился.

### Исправленный код

Поскольку оригинальный код представляет собой логическое месиво, чинить его костылями нет смысла. Вот правильная, чистая и понятная реализация классического алгоритма поиска O(n×m), которая делает ровно то, что от неё просят:
```Python

def find_substring(string: str, substring: str) -> int:
    """
    Funkce hledá podřetězec (substring) v řetězci (string)...
    """
    if len(substring) > len(string):
        return -1

    # Если подстрока пустая, по правилам возвращается 0
    # (хотя в условии написано neprázdný, лучше перестраховаться)
    if not substring:
        return 0

    # Перебираем возможные индексы начала подстроки в строке
    # Нам не нужно идти до самого конца строки, если оставшихся букв меньше,
    # чем длина искомого слова (отсюда len(string) - len(substring) + 1)
    for i in range(len(string) - len(substring) + 1):
        match = True

        # Проверяем посимвольно совпадает ли слово
        for j in range(len(substring)):
            if string[i + j] != substring[j]:
                match = False
                break # Если буква не совпала, прерываем внутренний цикл

        # Если внутренний цикл прошел до конца и match остался True
        if match:
            return i # Возвращаем индекс начала совпадения

    # Если мы всё перебрали и ничего не нашли
    return -1
```
